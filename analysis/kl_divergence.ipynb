```python
import pandas as pd
import matplotlib.pyplot as plt

from liars.constants import KL_PATH
from liars.utils import prefixes
```

```python
for prefix in prefixes.keys():
    P1_PATH = f"{KL_PATH}/{prefix}_p1.csv"
    p1_df = pd.read_csv(P1_PATH)
    fig, ax = plt.subplots(figsize=(5, 3))
    ax.boxplot(
        [p1_df[col] for col in p1_df.columns],
        tick_labels=p1_df.columns,
        showfliers=False
    )
    ax.set_title(f'KL Divergence for Pattern 1 \n(Prefix: {prefix} vs IT model)')
    ax.set_ylabel('KL Divergence')
    ax.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```
